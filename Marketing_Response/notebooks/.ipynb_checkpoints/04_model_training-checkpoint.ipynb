```python
def calculate_distribution_overlap(class0, class1, bins=100):
    """
    Вычисляет процент перекрытия двух распределений
    
    Параметры:
        class0, class1 (array-like): распределения для сравнения
        bins (int): количество бинов для гистограммы
        
    Возвращает:
        float: процент перекрытия (0-1)
    """
    hist0, edges = np.histogram(class0, bins=bins, density=True)
    hist1, _ = np.histogram(class1, bins=edges, density=True)
    return np.minimum(hist0, hist1).sum() / max(hist0.sum(), hist1.sum())

def plot_multiple_models_histograms(trained_models, X_train, y_train, X_val, y_val, 
                                   bins=30, figsize=None, models_to_plot=None):
    """
    Строит совмещенные гистограммы распределения скоров для нескольких моделей
    
    Параметры:
        trained_models (dict): Словарь с обученными моделями {имя_модели: модель}
        X_train, y_train (array-like): Обучающая выборка
        X_val, y_val (array-like): Валидационная выборка  
        bins (int): Количество бинов для гистограмм
        figsize (tuple): Размер полотна
        models_to_plot (list): Список имен моделей для отображения
    """
    if models_to_plot is None:
        models_to_plot = list(trained_models.keys())
    
    n_models = len(models_to_plot)
    if figsize is None:
        figsize = (6 * n_models, 10)
        
    fig, axes = plt.subplots(2, n_models, figsize=figsize)
    if n_models == 1:
        axes = axes.reshape(2, 1)
    
    def plot_single_hist(ax, y_true, y_proba, title):
        scores_class0 = y_proba[y_true == 0]
        scores_class1 = y_proba[y_true == 1]
        
        sns.histplot(scores_class0, color='red', label='Class 0', 
                    bins=bins, kde=True, alpha=0.5, ax=ax)
        sns.histplot(scores_class1, color='green', label='Class 1', 
                    bins=bins, kde=True, alpha=0.5, ax=ax)
        
        ax.set_title(title, fontsize=12)
        ax.set_xlabel('Вероятность класса 1', fontsize=10)
        ax.set_ylabel('Количество', fontsize=10)
        ax.legend(loc='upper center')
        ax.grid(True, linestyle='--', alpha=0.5)
        
        overlap = calculate_distribution_overlap(scores_class0, scores_class1)
        ax.text(0.95, 0.95, f'Перекрытие: {overlap:.1%}',
                ha='right', va='top', transform=ax.transAxes,
                bbox=dict(facecolor='white', alpha=0.7))
    
    for i, model_name in enumerate(models_to_plot):
        model = trained_models[model_name]
        
        try:
            y_proba_train = model.predict_proba(X_train)[:, 1]
            y_proba_val = model.predict_proba(X_val)[:, 1]
            
            plot_single_hist(axes[0, i], y_train, y_proba_train, 
                            f'{model_name} (Train)')
            plot_single_hist(axes[1, i], y_val, y_proba_val, 
                            f'{model_name} (Validation)')
        except Exception as e:
            print(f"Ошибка при обработке модели {model_name}: {str(e)}")
            continue
    
    plt.tight_layout()
    plt.show()
```

```python
plot_multiple_models_histograms(
    trained_models, 
    X_train, y_train, 
    X_val, y_val,
    models_to_plot=["Logistic Regression", "CatBoost", "SVM - Linear Kernel"]  
)
```

Интерпретация перекрытия - чем меньше перекрытие - тем лучше моделб различает классы.

Logistic Regression и CatBoost показывают одинаковое перекрытие относительно хорошие результаты. Это означает, что обе модели имеют схожую разделяющую способность на валидационных данных
